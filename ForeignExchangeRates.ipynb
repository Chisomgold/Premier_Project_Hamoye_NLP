```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_profiling as pp
```

```python
#reading dataset
fxr = pd.read_csv('Foreign_Exchange_Rates.csv')
```

# Exploring the dataset

```python
fxr.describe(include='all')
```

```python
#general overview of null values and datatypes
fxr.info()
```

```python
#glimpsing at data content
fxr.head(10)
```

```python
#Converting Time Series column to the datetime datatype
fxr['Time Serie'] = pd.to_datetime(fxr['Time Serie'])
```

```python
#Confirming changes to date column
fxr.info()
```

```python
#dropping unnecessary index column
fxr = fxr.drop(['Unnamed: 0'], axis=1)
```

```python
#Converting all columns except date to numerical values by force...
cols = fxr.columns.drop('Time Serie')
fxr[cols] = fxr[cols].apply(pd.to_numeric, errors='coerce')
```

```python
#confirming changes to columns datatype
fxr.info()
```

#### Notice that the columns now have null values compared to before.

```python
#null values examination
fxr.isnull().sum()
```

```python
#null values in percentage per column
fxr.isnull().mean() * 100
```

#### Null values are less than 5% of the entire dataset. Decision: drop all

```python
#dropping null values from the dataset
fxr = fxr.dropna()
```

```python
#reconfirming the state of null values in the data
fxr.isnull().sum()
```

```python
#reconfirming the datatypes in the dataset
fxr.info()
```

```python
#saving cleaned data to a new csv file
fxr.to_csv('Cleaned_Foreign_Exchange_Rates.csv', index=False)
```
